```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sb
import numpy as np
```

```python
# read dataset
#Planilha excel contendo a vendas dos produtos em colunas. Cada produto uma coluna.
df_table = pd.read_excel('/home/rafael/Downloads/projeto-aplicado-dados-produto-pivot.xlsx')

#df.head()
```

```python
df = pd.pivot_table(data=df_table,index=['PERIODO'],
                     columns=["NM_PRODUTO"],
                     aggfunc={"QTD_VENDA":np.sum})
```

```python
# get correlations
df_corr = df.corr()

df_corr.head()
```

```python
numeric_columns = df.columns

high_corr = [ ]

for c1 in numeric_columns:
  for c2 in numeric_columns:
    if c1 != c2 and c2 not in high_corr and df_corr[c1][c2] > 0.9:
      high_corr.append(c1)
```

```python
# map features to their absolute correlation values
df_corr = df_corr.corr().abs()

# set equality (self correlation) as zero
df_corr[df_corr == 1] = 0

# of each feature, find the max correlation
# and sort the resulting array in ascending order
corr_cols = df_corr.max().sort_values(ascending=False)

# display the highly correlated features
display(corr_cols[corr_cols > 0.8])
```

```python
np.ones_like(df_corr, dtype=np.bool)

mask = np.triu(np.ones_like(df_corr, dtype=np.bool))

fig, ax = plt.subplots(figsize=(24, 12))

# mask
mask = np.triu(np.ones_like(df_corr, dtype=np.bool))

# adjust mask and df
mask = mask[1:, :-1]
corr = df_corr.iloc[1:,:-1].copy()

# plot heatmap
sb.heatmap(corr, mask=mask, annot=True, fmt=".2f", cmap='Blues',
           vmin=-1, vmax=1, cbar_kws={"shrink": .8})

# yticks
plt.yticks(rotation=0)

plt.xticks(rotation=90)

plt.show()
```

```python
fig, ax = plt.subplots(figsize=(20, 16))

# mask
mask = np.triu(np.ones_like(df_corr, dtype=np.bool))

# adjust mask and df
mask = mask[1:, :-1]
corr = df_corr.iloc[1:,:-1].copy()

# color map
cmap = sb.diverging_palette(0, 230, 90, 60, as_cmap=True)

# plot heatmap
sb.heatmap(corr, mask=mask, annot=True, fmt=".2f", 
           linewidths=5, cmap=cmap, vmin=-1, vmax=1, 
           cbar_kws={"shrink": .8}, square=True)

# ticks
yticks = [i.upper() for i in corr.index]
xticks = [i.upper() for i in corr.columns]
plt.yticks(plt.yticks()[0], labels=yticks, rotation=0)
plt.xticks(plt.xticks()[0], labels=xticks)

# title
title = 'Matrix de correlação'
plt.title(title, loc='left', fontsize=18)

plt.show()
```
